# Cell 1: Imports 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import yaml
```

# Cell 2: Load Data

```python
with open("../config.yaml", "r") as f:
    config = yaml.safe_load(f)
df = pd.read_csv("../" + config["data"]["raw_train_path"], header=None, names=config["data"]["columns"])
```

# Cell 3: Initial EDA

```python
print(df.info())
print(df['label'].value_counts().head(10))
```

# Cell 4: Visualize Class Distribution

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='protocol_type', hue='label', data=df[df['label'].isin(['normal', 'neptune', 'satan'])])
plt.title("Protocol Type vs Top Attack Profiles")
plt.show()
```
